```python
import pandas as pd

ttb_id = '03211001000018'

master_df = pd.read_csv("dataset/master.csv", dtype=str)

filtered = master_df[master_df["TTB ID"] == ttb_id]

ground_truth = {
    "brand_name": filtered["Brand Name"].values[0] if not filtered.empty else None,
    "fanciful_name": filtered["Fanciful Name"].values[0] if not filtered.empty else None,
    # "class_type": filtered["Class/Type Desc"].values[0] if not filtered.empty else None,
    "alcohol_content": filtered["alcohol_content"].values[0] if not filtered.empty else None,
    "net_contents": filtered["net_contents"].values[0] if not filtered.empty else None,
}

ground_truth
```

```python
from ocr import ocr_image

image_path = f"dataset/images/{ttb_id}.jpg"
print(f'Using image: {image_path}')

text = ocr_image(image_path)
print(text)
```

```python
import re

from ocr import evaluate_ocr_text
```

```python
result = evaluate_ocr_text(
    text,
    **ground_truth
)

result
```

```python
import pandas as pd

df = pd.read_csv('dataset/master.csv')
```

```python
# filtered_full = df[df["net_contents"].notna() & df["alcohol_content"].notna()]

# Get the cleanest labels for initial testing
filtered_full = df[
    df["net_contents"].notna() &
    df["alcohol_content"].notna() &
    (df["net_contents"].astype(str).str.strip() != "") &
    (df["alcohol_content"].astype(str).str.strip() != "") &
    (df["Brand Name"].astype(str).str.strip() != "") &
    (df["image_path"].astype(str).str.strip() != "") &
    (df["fetch_status"] == "ok") &
    (df["variant_type"] == "original")
].copy()

filtered_full.to_csv('./dataset/master_filtered.csv', index=False)
```
